```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
N = 100000
df = 10  #degree of freedom
# Loi de student à 10 degrès de liberté 
T = np.random.standard_t(df, N)

# Loi normale centrée réduite 
Z = np.random.normal(0, 1, N)

# Loi du chi_2
X = np.random.chisquare(df, N)

T_proxi = Z / np.sqrt(X / df)

plt.figure(figsize=(12,8))
sns.histplot(T, color='#E67E22', alpha=0.8, stat='density', bins=50, label = f'Loi de Student à {df} degrès de liberté')
sns.kdeplot(T, color='black')
sns.histplot(T_proxi, color='#B5EA1C', alpha=0.3, stat='density', bins=50, label = f'Loi de Student approchée {df} degrès de liberté')
sns.kdeplot(T_proxi, color='black')
plt.legend()
plt.show()
```
