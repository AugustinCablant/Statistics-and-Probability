```python
import numpy as np 
import matplotlib.pyplot as plt 
```

$ X_1, ..., X_n | \theta \sim \mathcal{N}(\theta,1)$ 

$ \theta \sim \mathcal{N}(0,1)$ (densité à priori)

On montre faceilement que la loi a posteriori de $ \theta | X = (X_1, ..., X_n)$ est $ \mathcal{N}(\frac{n}{n+1} \bar{X_n}, \frac{1}{n+1})$. 

```python
n = 1000000
N = np.random.normal(0,1, size = n)
plt.hist(N, density = True, bins = 200, color = 'red', label = "loi a priori")
plt.legend()
plt.show()
```
